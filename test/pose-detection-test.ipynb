```python
import cv2
import numpy as np

# Load your image
image = cv2.imread("./resources\pose-img.jpg")
# Iterate through JSON data and draw bounding boxes and keypoints
for entry in results:
    box = entry["box"]
    keypoints = entry["keypoints"]

    x1, y1, x2, y2 = (
        int(box["x1"] * image.shape[1]),
        int(box["y1"] * image.shape[0]),
        int(box["x2"] * image.shape[1]),
        int(box["y2"] * image.shape[0])
    )

    # Draw bounding box
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green rectangle

    # Draw keypoints
    for x, y, vis in zip(keypoints["x"], keypoints["y"], keypoints["visible"]):
        x_pixel = int(x * image.shape[1])
        y_pixel = int(y * image.shape[0])

        if vis > 0.5:
            cv2.circle(image, (x_pixel, y_pixel), 5, (255, 0, 0), -1)  # Blue circle

# Show the image with bounding boxes and keypoints
cv2.imshow("Image with Bounding Boxes and Keypoints", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import os

img = cv2.imread('temp_file.jpg')
path = "./sample"
if not os.path.exists(path):
    os.mkdir(path)

for result in results:
    boxes = result.boxes.cpu().numpy()
    for i, box in enumerate(boxes):
        r = box.xyxy[0].astype(int)
        crop = img[r[1]:r[3], r[0]:r[2]]
        filename = str(i) + ".jpg"
        if (box.conf[0] > 0.5):
            filename = os.path.join(path, filename)
        cv2.imwrite(filename, crop)
```
